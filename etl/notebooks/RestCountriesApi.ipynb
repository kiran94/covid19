```python
import requests
import pandas as pd
from pprint import pprint
```

```python
url = 'https://restcountries.eu/rest/v2/all'
```

```python
result = requests.get(url).json()
```

```python
frame = pd.DataFrame.from_dict(result)
frame.rename(columns={ 'alpha2Code': 'iso3', 'alpha3Code': 'iso3', 'name': 'Country_Region' }, inplace=True)
frame.set_index('Country_Region', inplace=True)
```

```python
frame = pd.DataFrame.from_dict(result)

frame.rename(columns={ 
    'alpha2Code': 'iso2', 
    'alpha3Code': 'iso3', 
    'name': 'Country_Region' }, inplace=True)
frame.set_index('Country_Region', inplace=True)
```
